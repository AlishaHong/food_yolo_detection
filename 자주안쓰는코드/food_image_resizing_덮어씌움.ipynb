```python
import os
import cv2

# 이미지 크기를 변경할 경로 설정
train_image_folder = "kimchiSoup"
val_image_folder = "food"

# 이미지 크기를 640x640으로 변경하고 덮어씌우는 함수 (하위 폴더 포함)
def resize_images_in_folder(folder_path, size=(512, 512)):
    for root, dirs, files in os.walk(folder_path):  # os.walk()로 하위 폴더까지 모두 탐색
        for file in files:
            # 확장자가 이미지 파일인지 확인 (jpg, jpeg, png)
            if file.endswith(('.jpg', '.jpeg', '.JPG', '.png')):
                img_path = os.path.join(root, file)
                print(img_path)
                img = cv2.imread(img_path)
                
                if img is not None:
                    # 이미지 크기 조정
                    resized_img = cv2.resize(img, size)
                    # 이미지 덮어쓰기
                    cv2.imwrite(img_path, resized_img)
                    print(f"Resized and saved: {img_path}")
                else:
                    print(f"Failed to load image: {img_path}")

# Training과 Validation 폴더에서 이미지 크기 조정
resize_images_in_folder(train_image_folder)
resize_images_in_folder(val_image_folder)

print("모든 이미지가 512로 변경되었습니다.")
```
