```python
import os

# 클래스 인덱스 매핑 정의
class_names = [
    "갈비탕", "갈치조림", "감자탕", "감자튀김(스틱형)", "감자튀김(웨지감자)", "곱창전골", "김밥", "김치볶음밥", "깻잎장아찌", "꽁치조림", "꽃게탕",
    "녹두빈대떡", "달걀말이", "닭갈비", "닭볶음탕", "닭죽", "닭튀김", "돈가스", "돼지갈비", "두부구이", "두부김치", "떡갈비", "떡만둣국", "떡볶이",
    "라볶이", "류산슬덮밥", "매운탕", "문어숙회", "물회", "미소된장국", "병어찜", "부대찌개", "삼계탕", "삼선볶음밥", "삼선짬뽕", "새우볶음밥",
    "새우튀김", "소고기김밥", "소고기무국", "수제비", "순대국", "순대볶음", "순두부찌개", "시금치나물", "시래기된장국", "쌀밥",
    "알밥", "알탕", "양념게장", "어묵볶음", "영양돌솥밥", "오곡밥", "오므라이스(사각배달)", "오므라이스(접시)", "오징어볶음", "오징어젓갈", "육개장", "일식우동",
    "자장면", "잡탕밥", "전주비빔밥", "전주콩나물국밥", "제육덮밥", "쫄면", "참치마요삼각김밥", "채소튀김", "총각김치", "치킨윙", "카레라이스",
    "코다리조림", "콘스프", "콩국수", "콩밥", "크림소스스파게티", "토마토소스스파게티", "파전", "해물찜", "햄버거스테이크", "호박죽", "회냉면", "회덮밥", "훈제연어", "훈제오리"
]

# 클래스 이름과 인덱스를 매핑한 딕셔너리 생성
class_to_index = {name: idx for idx, name in enumerate(class_names)}
print(f"클래스-인덱스 매핑: {class_to_index}")

def update_last_line_index(txt_file_path, new_index):
    # txt 파일의 마지막 줄 인덱스를 새로운 값으로 수정
    with open(txt_file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # 마지막 줄의 첫 번째 인덱스를 새로운 값으로 변경
    if lines:
        parts = lines[-1].strip().split()  # 마지막 줄 분리
        if len(parts) > 0:  # 비어있지 않다면
            parts[0] = str(new_index)  # 인덱스 수정
            lines[-1] = ' '.join(parts) + '\n'  # 수정된 줄 저장

    # 수정된 내용 덮어쓰기
    with open(txt_file_path, 'w', encoding='utf-8') as f:
        f.writelines(lines)
    print(f"{txt_file_path} 수정 완료 - 마지막 줄 인덱스: {new_index}")

def update_classes_and_txt_files(base_label_folder):
    # 주어진 폴더 내 모든 클래스에 대해 txt 파일의 마지막 줄 인덱스를 업데이트
    for class_name, class_index in class_to_index.items():
        class_label_path = os.path.join(base_label_folder, class_name)

        if not os.path.exists(class_label_path):
            print(f"[경고] 라벨 폴더가 없습니다: {class_label_path}")
            continue

        # 해당 클래스 폴더의 모든 txt 파일 수정
        for txt_file in os.listdir(class_label_path):
            if txt_file.endswith('.txt'):
                txt_file_path = os.path.join(class_label_path, txt_file)
                update_last_line_index(txt_file_path, class_index)

# 경로 설정
train_label_folder = "C:/Users/SBA/repository/image_detection/음식이미지/Training/label"
valid_label_folder = "C:/Users/SBA/repository/image_detection/음식이미지/Validation/label"

# Training과 Validation 폴더의 라벨 파일 인덱스 업데이트 수행
update_classes_and_txt_files(train_label_folder)
update_classes_and_txt_files(valid_label_folder)

print("\n모든 클래스와 TXT 파일의 마지막 줄 인덱스가 업데이트되었습니다!")




# class_names = [
#     "Galbitang", "Galchi_Jorim", "Gamjatang", "French_Fries_Stick", 
#     "French_Fries_Wedge", "Gopchang_Jeongol", "Gimbap", "Kimchi_Fried_Rice", 
#     "Pickled_Perilla_Leaves", "Kkongchi_Jorim", "Flower_Crab_Soup", 
#     "Nokdu_Bindaetteok", "Rolled_Omelet", "Dakgalbi", "Dakbokkeumtang", 
#     "Chicken_Porridge", "Fried_Chicken", "Tonkatsu", "Pork_Ribs", 
#     "Grilled_Tofu", "Tofu_Kimchi", "Tteokgalbi", "Tteok_Mandu_Soup", 
#     "Tteokbokki", "Rabokki", "Rice_with_Lyusanseul", "Maeuntang", 
#     "Octopus_Sashimi", "Cold_Raw_Fish_Soup", "Miso_Soup", "Byeongeo_Jjim", 
#     "Budae_Jjigae", "Samgyetang", "Three_Delicacy_Fried_Rice", 
#     "Seafood_Jjamppong", "Shrimp_Fried_Rice", "Shrimp_Tempura", 
#     "Beef_Gimbap", "Beef_Radish_Soup", "Sujebi", "Sundae_Soup", 
#     "Stir_Fried_Sundae", "Sundubu_Jjigae", "Spinach_Namul", 
#     "Dried_Radish_Soybean_Paste_Soup", "Steamed_Rice", "Al_Bap", 
#     "Altang", "Spicy_Marinated_Crab", "Stir_Fried_Fish_Cake", 
#     "Nutritional_Stone_Pot_Rice", "Five_Grain_Rice", 
#     "Omelet_Rice_Square", "Omelet_Rice_Plate", "Stir_Fried_Squid", 
#     "Fermented_Squid", "Yukgaejang", "Japanese_Udon", "Jajangmyeon", 
#     "Mixed_Seafood_Rice", "Jeonju_Bibimbap", "Jeonju_Bean_Sprout_Soup", 
#     "Spicy_Pork_Rice", "Cold_Spicy_Noodles", "Tuna_Mayo_Triangle_Gimbap", 
#     "Vegetable_Tempura", "Chonggak_Kimchi", "Chicken_Wings", 
#     "Curry_Rice", "Braised_Pollock", "Corn_Soup", "Kongguksu", 
#     "Mixed_Grain_Rice", "Cream_Sauce_Spaghetti", "Tomato_Sauce_Spaghetti", 
#     "Pajeon", "Haemul_Jjim", "Hamburger_Steak", "Pumpkin_Porridge", 
#     "Hoe_Naengmyeon", "Raw_Fish_Bibimbap", "Smoked_Salmon", 
#     "Smoked_Duck"
# ]
```
