```python
import os
import json

def change_label_in_json(folder_path, new_label):
    #     
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".json"):  # JSON  
            file_path = os.path.join(folder_path, file_name)
            
            # JSON  
            with open(file_path, 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
            
            # label  
            if "shapes" in data:
                for shape in data["shapes"]:
                    if "label" in shape:
                        shape["label"] = new_label  #   
            
            #   
            with open(file_path, 'w', encoding='utf-8') as json_file:
                json.dump(data, json_file, ensure_ascii=False, indent=4)
            print(f"Updated {file_name}")

#     label 
folder_path = "C:/Users/Sesame/Desktop/6mix_scaling_images/1st_2nd_mix_scaled/04011011"  # JSON    
new_label = "04011011"  #  label 

#  
change_label_in_json(folder_path, new_label)
```
